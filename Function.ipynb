```python
def getName(name):
    stack = []
    lname, fname = name.split(", ")
    lname = [lname]
    for n in fname.split(" "):
        stack.append(n)
    return stack+lname
```

```python
def getInitials(name):
    name = getName(name)
    if len(name)==0:
        return None
    name = [n[0] for n in name]
    
    def combinations(l):
        if l:
          result = combinations(l[:-1])
          return result + [c + [l[-1]] for c in result]
        else:
          return [[]]

    res = [n for n in combinations(name) if name[0] in n and name[-1] in n]
    for i,r in enumerate(res):
        temp = res.pop(i)
        if r not in res:
            res.insert(i,r)
    final = []
    for r in res:
        final.append("".join(r))
    return final  
    
```

```python
def sort(arr):
    n = len(arr)
    for i in range(1, n):
        temp = arr[i]
        j = i - 1
        while j >= 0 and len(temp) > len(arr[j]):
            arr[j + 1] = arr[j]
            j -= 1
 
        arr[j + 1] = temp
```
